### Mapping dictionary from image_id to dialogs

#### for val

```python
val_dialog_path = "/hub_data3/miso/datasets/visdial/dialogs/visdial_1.0_val.json"
f = open(val_dialog_path)
val_dialog = json.load(f)
```

```python
val_answers_list = val_dialog['data']['answers']
val_questions_list = val_dialog['data']['questions']
val_questions_list = [question + '?' for question in val_questions_list]
```

```python
val_dialog_map = dict()
for sample in val_dialog['data']['dialogs']:
    image_id = sample['image_id']
    caption = sample['caption']
    dialogs = sample['dialog']
    val_dialog_map[image_id] = dict()
    val_dialog_map[image_id]['dialogs'] = list()
    for dialog in dialogs:
        answer_idx = dialog['answer']
        question_idx = dialog['question']
        val_dialog_map[image_id]['dialogs'].append({'answer': val_answers_list[answer_idx], 'question': val_questions_list[question_idx]})
    val_dialog_map[image_id]['caption'] = caption
val_dialog_map
```

```python
with open('/hub_data3/miso/datasets/visdial/dialogs/val_imgid_dialogs.json', 'w') as f:
    json.dump(val_dialog_map, f)
```

### Visualize images

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

# val set
val_folder_path = '/hub_data3/miso/datasets/visdial/images/VisualDialog_val2018/' 
val_file_list = os.listdir(val_folder_path)
```

```python
val_imgid_map = {}
for file_list in val_file_list:
    image_id = int(file_list.split('_')[-1].split('.')[0])
    val_imgid_map[image_id] = file_list
```

```python
imgid = 88394
file_path = os.path.join(val_folder_path, val_imgid_map[imgid])
image_pil = Image.open(file_path)
image = np.array(image_pil)

plt.imshow(image)
plt.show()
```
